```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('electronics.csv')
dataset.head(10)
```

```python
dataset.tail(20)
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
from datetime import datetime
pd.to_datetime(dataset['timestamp'])
```

```python
dataset['category'] = dataset['category'].astype(str)
```

```python
dataset['rating'] = dataset['rating'].astype(float) 
```

```python
dataset['user_id'] = dataset['user_id'].astype(str)
```

```python
dataset['item_id'] = dataset['item_id'].astype(str)
```

```python
dataset.describe()
```

```python
dataset.nunique()
```

```python
dataset.duplicated().sum()
```

```python
dataset.isnull().sum()
```

```python
dataset['rating'].value_counts()
```

```python
dataset['year'] = pd.DatetimeIndex(dataset['timestamp']).year
dataset['year'].value_counts()
```

```python
dataset['month'] = pd.DatetimeIndex(dataset['timestamp']).month
dataset['month'].value_counts()
```

```python
dataset.dropna(inplace=True)
dataset.isnull().sum()
```

```python
dataset['rating'].value_counts()
```

```python
sns.countplot(x='rating', data=dataset)
```

```python
sns.countplot(x='year', data=dataset)
```

```python
sns.countplot(x='month', data=dataset)
```

```python
sns.countplot(x='gender', data=dataset, order=dataset['gender'].value_counts().iloc[-3:].index)
```

```python
sns.countplot(x='gender', data=dataset[dataset['year'] == 2013], order=dataset['gender'].value_counts().iloc[1:3].index)
```

```python
sns.countplot(x='gender', data=dataset[dataset['year'] == 2014], order=dataset['gender'].value_counts().iloc[1:3].index)
```

```python
sns.countplot(x='gender', data=dataset[dataset['year'] == 2015], order=dataset['gender'].value_counts().iloc[1:3].index)
```

```python
sns.countplot(x='category', data=dataset[dataset['year'] == 2016], order=dataset['category'].value_counts().iloc[1:4].index)
```

```python
sns.countplot(x='category', data=dataset[dataset['year'] == 2017], order=dataset['category'].value_counts().iloc[1:4].index)
```

```python
dataset['gender'].value_counts(normalize=True)
dataset.groupby('gender')['rating'].count().sort_values(ascending=True).head().plot(kind='pie')
sns.set_style('white')
```

```python
dataset['category'].value_counts(normalize=True)
dataset.groupby('category')['rating'].count().sort_values(ascending=True).head().plot(kind='pie')
sns.set_style('white')
```

```python
dataset['rating'].value_counts(normalize=True)
dataset.groupby('rating')['year'].count().sort_values(ascending=True).head().plot(kind='pie')
sns.set_style('white')
```

```python
dataset['year'].value_counts(normalize=True)
dataset.groupby('year')['rating'].count().sort_values(ascending=True).head().plot(kind='pie')
sns.set_style('white')
```
